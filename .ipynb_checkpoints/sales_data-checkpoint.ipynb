## Medium

## 1. What are the three most popular categories, by total sales?

```python
quantity = df['quantity']
price = df['price']
quantity*price
df2 = df.assign(product_price=quantity*price)
print((df2.groupby(by=['category']).sum()['product_price']).sort_values(ascending=False))
print("The three most popular categories, by total sales, are Clothing, Shoes, and Technology")
```

## 2. What are the total sales attributed to customers over the age of 45?

```python
mdf = pd.merge(jdf, df, left_on="id", right_on="customer_id")
quantity = mdf['quantity']
price = mdf['price']
ndf = mdf.assign(product_price=quantity*price)
odf = ndf[ndf['age']>45]
print("The total sales attributed to customers over the age of 45 is " + str(odf['product_price'].sum()) + ".")
```

## 3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)

```python
mdf['invoice_date'] = pd.to_datetime(mdf['invoice_date'])
invoices_per_month = mdf.set_index('invoice_date').resample('M').size()
plt.figure(figsize=(12, 6))
plt.plot(invoices_per_month.index, invoices_per_month)
plt.xlabel('Month')
plt.ylabel('Number of Invoices')
plt.title('Number of Invoices Per Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Hard

## 1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth). 

```python
mdf['age_range'] = (mdf['age'] // 10 * 10).astype(str) + '-' + ((mdf['age'] // 10 * 10) + 9).astype(str)
mdf['product_price'] = mdf['quantity'] * mdf['price']
pivot_table = mdf.pivot_table(values='product_price', index='category', columns='age_range', aggfunc='sum', fill_value=0)
print(pivot_table)
pivot_table.plot(kind='bar', figsize=(14, 7))
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Breakdown of Sales by Category and Decade Age Range')
plt.xticks(rotation=45)
plt.legend(title='Age Range')
plt.tight_layout()
plt.show()
```
